Class and Objects

```python
class Student:
    name = "Tejaswini Suman"

s1 = Student()
print(s1.name)

s2 = Student()
print(s2.name)
```

```python
class Car:
    color = "Black"
    brand = "Land Rover"
    model = "Defender"
    Engine_Type = "V8"

c1 = Car()
print(f"My dream car is a {c1.color} {c1.brand} {c1.model} {c1.Engine_Type}")
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        

s1 = Student("Abhinav Kashyap", "8.5")
print(s1.name, s1.marks)


s2 = Student("Tejaswini Kashyap", "8.5")
print(s2.name, s2.marks)
    
```

Class and Instance Attributes

```python
class Student:
    university_name = "Pimpri Chinchwad University"
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        

s1 = Student("Abhinav Kashyap", "8.5")
print(s1.name, s1.marks)
print(s1.university_name)

print("\n")

s2 = Student("Tejaswini Kashyap", "8.5")
print(s2.name, s2.marks)
print(s2.university_name)
```

Methods

```python
class Student:
    university_name = "Pimpri Chinchwad University"
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    def welcome(self):
        print("Welcome Student", self.name)

    def get_marks(self):
        return self.marks
        

s1 = Student("Abhinav Kashyap", "8.5")
print(s1.name, s1.marks)
print(s1.university_name)

print("\n")

s2 = Student("Tejaswini Kashyap", "8.5")
print(s2.name, s2.marks)
print(s2.university_name)

print("\n")

s1.welcome()
s2.welcome()

print("\n")

print(s1.get_marks())
print(s2.get_marks())
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name 
        self.marks = marks

    def get_average(self):
        sum = 0
        for val in self.marks:
            sum = val + sum
        print("Hi", self.name, "your avg score is:", sum//3 )    
       
        

n1 = Student("Abhinav Kashyap",[67,92,76])        
print(n1.get_average())


        
```
