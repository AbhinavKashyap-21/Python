# Patterns

Pattern 1

```python
def pattern1(n):
    for i in range(0, n):
        print("*" * n)
        
num = int(input("Enter the number: "))

pattern1(num)
```

Pattern 4

```python
def pattern4(n):
    for i in range(n,0,-1):
        for j in range(1, i+1):   
            print(j, end=" ")
        print()

num = int(input("Enter your number"))
pattern4(num)
```

```python
def pattern5(n):
    for i in range(1, n+1):
        for j in range(1, i+1):
            print(i, end=" ")
        print()

num = int(input("Enter your number: ")) 
pattern5(num)
```
