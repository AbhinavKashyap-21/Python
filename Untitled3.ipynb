# Serializtion and Deserialization 

using JSON

```python
import json

user_data = {
    "name": "Lancelot",
    "quest": "The  Holy Grail",
    "is_knight": True,
    "inventory" : ["Shield", "Sword", "Helmet"]
}

serialized_data = json.dumps(user_data)
print(f"Serialized String:{serialized_data}")
```

```python
original_object = json.loads(serialized_data)
print(f"Back to Object: {original_object['name']} is on a quest for {original_object['quest']}.")
```

using Pickle

```python
import pickle

class Knight:
    def __init__(self, name, colour):
        
        self.name = name
        self.colour = colour

Bahubali = Knight("Robin Hood", "Yellow")

serialized_data = pickle.dumps(Bahubali)
print(serialized_data)
```

```python
unserialized_data = pickle.loads(serialized_data)
print(f"Reconstructed: {unserialized_data.name}'s favorite color is {unserialized_data.colour}.")
```

```python
import json

# 1. Our Python Data
hero_stats = {
    "name": "Sir Lancelot",
    "level": 15,
    "inventory": ["Shield", "Bread", "Grail"],
    "is_alive": True
}

# 2. Open a new file called 'hero.json' in 'w' (write) mode
with open('hero.json', 'w') as f:
    # Use indent=4 to make the file look pretty and readable
    json.dump(hero_stats, f, indent=4)

print("File 'hero.json' has been created!")
```

```python
import json

# 1. Open the file in 'r' (read) mode
with open('hero.json', 'r') as f:
    # Load the data back into a Python variable
    loaded_hero = json.load(f)

# 2. Access the data to prove it's a Python object again
print(f"Hero Name: {loaded_hero['name']}")
print(f"First Item: {loaded_hero['inventory'][0]}")

if loaded_hero['is_alive']:
    print("The hero is ready for the next quest!")
```

```python
def set(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    print(f"Age set to {age}") 
try:
    set(-5)
except ValueError as e:
    print(e)
```

```python
class AgeError(Exception):
    pass

def set(age):
    if age < 0:
        raise AgeError("Age cannot be negative.")
    print(f"Age set to {age}")

try:
    set(-5)
except AgeError as e:
    print(e)
```
