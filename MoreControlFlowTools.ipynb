# Changing the size of the dictionary using For Loop

```python
users = {'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}

for user, status in users.copy().items():
    if status == 'inactive':
        del users[user]

        print(user)
```

```python
users = {'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}

active_users = {}
for user, status in users.items():
    if status == 'active':
        active_users[user] = status

        print(active_users)
```

# Break and Continue Statements 

The break statement breaks out of the innermost enclosing for or while loop:

```python
items = ["Candy", "Apple", "Golden Ticket", "Bat", "Ball"]

for i, item in enumerate(items):
    print(i,item)

    if item == "Golden Ticket":
        print("Braking out of the loop")
        break
print("Loop is finished")        
```

```python
items = ["Candy", "Apple", "Golden Ticket", "Bat", "Ball"]

for item in items:
    if item == "Apple":
        print("Not in the mood to eat apple")
        continue

    print(f"Checking: {item}")    

        
```

# Match Statements

match statement is a more powerful version of switch-case statements found in java or any other languages, unlike simple switch
that compare values match can look on the structure of data (like checking length of a list or keys in the dictionary)

```python
day = 4
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")

        
```

```python
day = 8
match day:
    case 1|2|3|4|5:
        print("Today is a weekday")
    case 6|7:
        print("I love weekends")
    case _:
        print("Happy New Year")

        
```

```python
month = 5
day = 8
match day:
    case 1|2|3|4|5 if month == 4:
        print("Weekday of April")
    case 1|2|3|4|5 if month == 5:
        print("Weekday of May")
    case _:    
        print("No Match")

        

        
```

# Defining Functions

```python
def fib(n):
    a,b = 0,1
    while a < n:
        print(a, end = " ") # with the use of end output will print on a single line not in multiple lines
        a,b = b, a+b
    print()

fib(2000)    
```

```python
def fib2(n):
    result = []
    a,b = 0,1
    while a < n:
        result.append(a)
        a,b = b, a+b
    return result
fb100 = fib2(100)
print(fb100)        
```

# Keyword and Positional Argument

Keyword arguments mean you pass values by parameter names while calling the function.

```python
def profile(name,age):
    print("My name is ", name)
    print("My age is ", age)

profile(name = "Suman", age = "20")    
profile(age = "21", name = "Abhinav")
```

# Lambda Functions

Lambda Functions are anonymous functions means that the function is without a name. As we already know def keyword is used to define a normal function in Python. Similarly, lambda keyword is used to define an anonymous function in Python. 

```python
n = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(n(5))
print(n(-10))
print(n(0))

```

```python
check = lambda x : "Even" if x % 2 == 0 else "odd"
print(check(4))
print(check(3))
```

```python
calc = lambda x,y: (x + y, x * y)
res = calc(3, 4)
print(res)
```

```python
n = [1,2,3,4,5,6]
even = filter(lambda x: x%2==0, n)
print(list(even))
```

```python
n = [1,2,3,4,5]
Triple = map(lambda x: x * 3, n)
print(list(Triple))
```
